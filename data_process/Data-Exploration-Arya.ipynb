```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.preprocessing import OneHotEncoder
```

```python
# input data/change column name/merge data
people_data=pd.read_csv('people.csv',parse_dates=['date'])
test_act_data=pd.read_csv('act_test.csv',parse_dates=['date'])
train_act_data=pd.read_csv('act_train.csv',parse_dates=['date'])
```

```python
def data_treatment(name):
    dataset=name
    for i in list(dataset.columns):
        if i not in ['people_id','activity_id','date','char_38','outcome']:
            if dataset[i].dtype=='object':
                dataset[i].fillna('type 0',inplace=True)
                dataset[i] = dataset[i].apply(lambda x: x.split(' ')[1])
                dataset[i]=pd.to_numeric(dataset[i].astype(np.int32))
            elif dataset[i].dtype=='bool':
                dataset[i]=dataset[i].astype(np.int8)
    dataset['people_id']=dataset['people_id'].apply(lambda x:x.split('_')[1])
    dataset['people_id']=pd.to_numeric(dataset['people_id']).astype(int)
    return dataset
# train_act_data=data_treatment(train_act_data)   
```

```python
train_act_data=train_act_data.drop('char_10',axis=1)
test_act_data=test_act_data.drop('char_10',axis=1)
```

```python
train_act_data=data_treatment(train_act_data)   
test_act_data=data_treatment(test_act_data)   
people_data=data_treatment(people_data)
```

```python
column_people=np.arange(1,41)
a=people_data.columns[column_people]
b=['pp_'+people_data.columns[i] for i in column_people]
mapping_people=dict(zip(a,b))

column_train=[2,*np.arange(4,14)]
a_1=train_act_data.columns[column_train]
b_1=['act_'+train_act_data.columns[i] for i in column_train]
mapping_train=dict(zip(a_1,b_1))

column_test=[2,*np.arange(4,13)]
a_2=test_act_data.columns[column_test]
b_2=['act_'+test_act_data.columns[i] for i in column_test]
mapping_test=dict(zip(a_2,b_2))
 
people_data=people_data.rename(columns=mapping_people)
test_act_data=test_act_data.rename(columns=mapping_test)
train_act_data=train_act_data.rename(columns=mapping_train)
```

```python
# test_data=pd.merge(test_act_data,people_data,on='people_id')
train_data=train_act_data.merge(people_data,on='people_id',how='left', left_index=True)
test_data=test_act_data.merge(people_data,on='people_id',how='left', left_index=True)
```
